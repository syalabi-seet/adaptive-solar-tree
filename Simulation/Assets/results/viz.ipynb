```python
import pandas as pd
import os

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r"F:\Projects\AdaptiveSolarTree\Simulation\Assets\results\thirdRun"

reward = pd.read_csv(os.path.join(path, "thirdRun_Etiolation_REWARD.csv"))[['Step', 'Value']]
reward
```

```python
steps = pd.read_csv(os.path.join(path, "thirdRun_Etiolation_STEPS.csv"))[['Step', 'Value']]
steps
```

```python
fig, ax = plt.subplots(2, 1, figsize=(5, 5), sharex=True)

ax[0].plot(reward['Step'], reward['Value'])
ax[0].set_ylabel("Cumulative reward")
ax[1].plot(steps['Step'], steps['Value'])
ax[1].set_ylabel("Episode steps")

plt.xlabel("Environment steps")
plt.tight_layout()
plt.show()
```
