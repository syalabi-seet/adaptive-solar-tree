```python
import pandas as pd
```

```python
bom_1 = pd.read_csv(r"F:\Projects\AdaptiveSolarTree\Models\Build\v4\LegoBuildv4.csv")[:-3][['BLItemNo', 'Qty']]
bom_1
```

```python
bom_2 = pd.read_csv(r"F:\Projects\AdaptiveSolarTree\Models\Build\v4\LegoBuildv5.csv")[:-3][['BLItemNo', 'Qty']]
bom_2
```

```python
new_bom = pd.merge(bom_2, bom_1, how="left", on="BLItemNo")
new_bom = new_bom.fillna(0)
new_bom['delta_Qty'] = new_bom['Qty_x'] - new_bom['Qty_y']
new_bom = new_bom.query("delta_Qty > 0")[['BLItemNo', 'delta_Qty']]
new_bom
```

```python
new_bom.to_csv("new_BOM.csv", index=False)
```
